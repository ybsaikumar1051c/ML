### 10. Implement K-Means Clustering Algorithm

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
X, _ = make_blobs(n_samples=700, centers=7, cluster_std=0.70, random_state=40)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], cmap='dark', s=100, alpha=0.5)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
kmeans = KMeans(n_clusters=7)
```

```python
kmeans.fit(X)
```

```python
y_kmeans = kmeans.predict(X)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=100, alpha=0.5)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')
plt.title('KMeans Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
data = np.random.rand(200, 2)
```

```python
kmeans = KMeans(n_clusters=5, random_state=50)
kmeans.fit(data)
```

```python
centers = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='k', s=100)
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=200, label='Centers')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
